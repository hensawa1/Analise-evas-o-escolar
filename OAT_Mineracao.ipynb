```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Carregar CSV
df_original = pd.read_csv("MICRODADOS_ENEM_ESCOLA.csv", sep=";", encoding="latin-1", low_memory=False)

# Colunas relevantes
colunas_relevantes = [
    'TP_DEPENDENCIA_ADM_ESCOLA', 'TP_LOCALIZACAO_ESCOLA',
    'NU_MEDIA_CN', 'NU_MEDIA_CH', 'NU_MEDIA_LP', 'NU_MEDIA_MT',
    'NU_MEDIA_RED', 'NU_MEDIA_OBJ', 'NU_MEDIA_TOT',
    'NU_TAXA_APROVACAO', 'NU_TAXA_REPROVACAO', 'NU_TAXA_ABANDONO',
    'NU_TAXA_PERMANENCIA'
]

df = df_original[colunas_relevantes].copy()

# Remover colunas com mais de 30% de NaNs
limite_nan = 0.3
colunas_validas = df.columns[df.isnull().mean() < limite_nan]
df = df[colunas_validas].dropna()

# Recolocar as colunas para classificar escola
for col in ['TP_DEPENDENCIA_ADM_ESCOLA', 'TP_LOCALIZACAO_ESCOLA']:
    if col not in df.columns:
        df[col] = df_original[col]

# Classificar tipo de escola
def classificar_escola(row):
    if row['TP_LOCALIZACAO_ESCOLA'] == 2:
        return 'Rural'
    elif row['TP_DEPENDENCIA_ADM_ESCOLA'] == 1:
        return 'Federal'
    elif row['TP_DEPENDENCIA_ADM_ESCOLA'] == 2:
        return 'Estadual'
    elif row['TP_DEPENDENCIA_ADM_ESCOLA'] == 4:
        return 'Privada'
    else:
        return 'Outros'

df['TIPO_ESCOLA'] = df.apply(classificar_escola, axis=1)
df = df[df['TIPO_ESCOLA'].isin(['Federal', 'Estadual', 'Privada', 'Rural'])]

# Normalização com MinMaxScaler por tipo de escola
minmax_scaler = MinMaxScaler()
df_normalizado = []

for tipo, grupo in df.groupby('TIPO_ESCOLA'):
    dados = grupo.drop(columns=['TP_DEPENDENCIA_ADM_ESCOLA', 'TP_LOCALIZACAO_ESCOLA', 'TIPO_ESCOLA'])
    if dados.shape[0] > 0:
        escalado = minmax_scaler.fit_transform(dados)
        df_escalado = pd.DataFrame(escalado, columns=dados.columns)
        df_escalado['TIPO_ESCOLA'] = tipo
        df_normalizado.append(df_escalado)

# Concatenar resultados
if df_normalizado:
    df_final = pd.concat(df_normalizado)
    media_por_tipo = df_final.groupby('TIPO_ESCOLA').mean() * 100

    # Gráfico
    plt.figure(figsize=(14, 6))
    media_por_tipo.T.plot(kind='bar', figsize=(16, 6), colormap='Set2', edgecolor='black')
    plt.title("Comparação Normalizada das Métricas por Tipo de Escola (MinMax %)", fontsize=14)
    plt.ylabel("Média Normalizada (%)")
    plt.xlabel("Indicadores Educacionais")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(title="Tipo de Escola", loc='upper right')
    plt.tight_layout()
    plt.show()
else:
    print("Nenhum dado válido para normalizar.")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Carrega o CSV
df = pd.read_csv(
    r'C:\Users\henr\Desktop\Projeto_educacao_evasao_escolar\Projeto_educacao_evasao_escolar\MICRODADOS_ENEM_ESCOLA.csv',
    sep=';',
    low_memory=False,
    encoding='latin-1'
)

# Define colunas
colunas_numericas = ['NU_MEDIA_OBJ', 'NU_MEDIA_MT', 'NU_MEDIA_LP']
target_column = 'NU_TAXA_ABANDONO'

# Verifica colunas
for col in colunas_numericas:
    if col not in df.columns:
        raise ValueError(f"Coluna '{col}' não encontrada no DataFrame.")
if target_column not in df.columns:
    raise ValueError(f"Coluna alvo '{target_column}' não encontrada no DataFrame.")

# Seleciona dados
X = df[colunas_numericas]
y = df[target_column]

# Binariza evasão
y = (y > 0).astype(int)

# Mostra valores ausentes
print("Valores ausentes por coluna:")
print(X.isnull().sum())

# Preenche valores ausentes
X = X.fillna(X.mean())
y = y.loc[X.index]

print(f"\nNúmero de escolas consideradas após tratamento: {len(X)}")

# Validação cruzada
kf = KFold(n_splits=5, shuffle=True, random_state=42)
acuracias = []

print("\nDesempenho por fold:")

for i, (train_idx, test_idx) in enumerate(kf.split(X), 1):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    modelo = LogisticRegression(max_iter=1000)
    modelo.fit(X_train, y_train)
    y_pred = modelo.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    acuracias.append(acc)

    print(f"Publica {i} (teste em 1/5 das escolas): {acc * 100:.2f}%")

# Média
print(f"\nAcurácia média (todos as escolas): {np.mean(acuracias) * 100:.2f}%")
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

# Carregar os dados
cols = [
    "TP_DEPENDENCIA_ADM_ESCOLA", "TP_LOCALIZACAO_ESCOLA",
    "NU_TAXA_APROVACAO", "NU_TAXA_REPROVACAO", "NU_TAXA_ABANDONO",
    "NU_MEDIA_TOT"
]
df = pd.read_csv("MICRODADOS_ENEM_ESCOLA.csv", encoding="latin-1", sep=";", usecols=cols)

# Remover linhas com valores ausentes
df = df.dropna()

# Separar X e y
X = df.drop("NU_MEDIA_TOT", axis=1)
y = (df["NU_MEDIA_TOT"] > df["NU_MEDIA_TOT"].median()).astype(int)

# Codificar variáveis categóricas
X = pd.get_dummies(X, columns=["TP_DEPENDENCIA_ADM_ESCOLA", "TP_LOCALIZACAO_ESCOLA"], drop_first=True)

# Seleção de atributos
selector = SelectKBest(score_func=mutual_info_classif, k='all')
X_new = selector.fit_transform(X, y)

# Exibir colunas selecionadas com pontuações
selected_columns = X.columns[selector.get_support()]
scores = selector.scores_[selector.get_support()]
for col, score in zip(selected_columns, scores):
    print(f"{col}: {score:.4f}")

# Dividir dados
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)

# Padronização
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Treinar o modelo
model = MLPClassifier(random_state=42, max_iter=500)
model.fit(X_train, y_train)

# Avaliar
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia do modelo: {accuracy:.4f}")


```
