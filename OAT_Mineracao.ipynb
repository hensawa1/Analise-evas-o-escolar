```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Colunas relevantes para o gráfico
colunas_taxas = [
    'NU_TAXA_APROVACAO', 'NU_TAXA_REPROVACAO',
    'NU_TAXA_ABANDONO', 'NU_TAXA_PERMANENCIA',
    'TP_DEPENDENCIA_ADM_ESCOLA', 'TP_LOCALIZACAO_ESCOLA'
]

# Inicializa leitura por blocos
amostra = []
limite = 10000
lidas = 0

for chunk in pd.read_csv(
    "MICRODADOS_ENEM_ESCOLA.csv",
    sep=";",
    encoding="latin-1",
    chunksize=1000,
    low_memory=False,
    on_bad_lines="skip"
):
    if not all(col in chunk.columns for col in colunas_taxas):
        continue

    chunk = chunk[colunas_taxas].dropna()
    if not chunk.empty:
        amostra.append(chunk)
        lidas += len(chunk)
        if lidas >= limite:
            break

# Verificação final
if not amostra:
    raise ValueError("❌ Nenhum dado válido foi encontrado nos chunks lidos.")

# Concatena e limita a 10.000 linhas
df = pd.concat(amostra).head(limite)

# Classificação por tipo de escola
def classificar_escola(row):
    if row['TP_LOCALIZACAO_ESCOLA'] == 2:
        return 'Rural'
    elif row['TP_DEPENDENCIA_ADM_ESCOLA'] == 1:
        return 'Federal'
    elif row['TP_DEPENDENCIA_ADM_ESCOLA'] == 2:
        return 'Estadual'
    elif row['TP_DEPENDENCIA_ADM_ESCOLA'] == 4:
        return 'Privada'
    return 'Outros'

df['TIPO_ESCOLA'] = df.apply(classificar_escola, axis=1)
df = df[df['TIPO_ESCOLA'].isin(['Federal', 'Estadual', 'Privada', 'Rural'])]

# Selecionar apenas as taxas
taxas = ['NU_TAXA_APROVACAO', 'NU_TAXA_REPROVACAO', 'NU_TAXA_ABANDONO', 'NU_TAXA_PERMANENCIA']

# Normalização com MinMaxScaler por tipo de escola
df_normalizado = []
scaler = MinMaxScaler()

for tipo, grupo in df.groupby('TIPO_ESCOLA'):
    dados = grupo[taxas]
    if not dados.empty:
        dados_normalizados = scaler.fit_transform(dados)
        df_norm = pd.DataFrame(dados_normalizados, columns=taxas)
        df_norm['TIPO_ESCOLA'] = tipo
        df_normalizado.append(df_norm)

# Concatenar dados normalizados
df_final = pd.concat(df_normalizado)
media_por_tipo = df_final.groupby('TIPO_ESCOLA').mean() * 100

# Gráfico final
plt.figure(figsize=(12, 6))
media_por_tipo.T.plot(kind='bar', figsize=(14, 6), colormap='Set2', edgecolor='black')
plt.title("📊 Comparação Normalizada das Taxas por Tipo de Escola", fontsize=14)
plt.ylabel("Taxa Média Normalizada (%)")
plt.xlabel("Indicadores Educacionais")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Tipo de Escola", loc='upper right')
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Carregar dados
df = pd.read_csv("MICRODADOS_ENEM_ESCOLA.csv", sep=';', encoding='latin-1', low_memory=False)

# Notas como entrada
notas = ['NU_MEDIA_CN', 'NU_MEDIA_CH', 'NU_MEDIA_LP', 'NU_MEDIA_MT', 'NU_MEDIA_RED']

# Taxas como alvo
taxas = ['NU_TAXA_APROVACAO', 'NU_TAXA_REPROVACAO', 'NU_TAXA_ABANDONO', 'NU_TAXA_PERMANENCIA']

# Função para classificar tipo de escola
def classificar_escola(row):
    if row['TP_LOCALIZACAO_ESCOLA'] == 2:
        return 'Rural'
    elif row['TP_LOCALIZACAO_ESCOLA'] == 1:
        if row['TP_DEPENDENCIA_ADM_ESCOLA'] == 1:
            return 'Federal'
        elif row['TP_DEPENDENCIA_ADM_ESCOLA'] == 2:
            return 'Estadual'
        elif row['TP_DEPENDENCIA_ADM_ESCOLA'] == 4:
            return 'Privada'
    return 'Outros'

# Aplicar classificação
df['TIPO_ESCOLA'] = df.apply(classificar_escola, axis=1)

# Tipos a analisar
tipos_escola = ['Federal', 'Estadual', 'Privada', 'Rural']

for tipo in tipos_escola:
    print(f"\n==================== 🏫 Tipo de Escola: {tipo} ====================")

    # Filtrar dados e remover nulos
    df_tipo = df[df['TIPO_ESCOLA'] == tipo]
    df_tipo = df_tipo[notas + taxas].dropna()

    if df_tipo.empty:
        print("⚠️ Sem dados suficientes.")
        continue

    X = df_tipo[notas]

    for taxa in taxas:
        y = df_tipo[taxa]
        y_bin = (y > y.median()).astype(int)

        print(f"\n Avaliando {taxa}: ")

        # Divisão treino/teste
        X_train, X_test, y_train, y_test = train_test_split(
            X, y_bin, test_size=0.2, random_state=42, stratify=y_bin
        )

        modelo = LogisticRegression(max_iter=1000)
        modelo.fit(X_train, y_train)
        y_pred = modelo.predict(X_test)
        acc_split = accuracy_score(y_test, y_pred)
        print(f"🎯 Acurácia (train_test_split): {acc_split * 100:.2f}%")

        # Validação cruzada com KFold
        kf = KFold(n_splits=5, shuffle=True, random_state=42)
        acuracias_kfold = []

        for train_idx, test_idx in kf.split(X):
            X_ktrain, X_ktest = X.iloc[train_idx], X.iloc[test_idx]
            y_ktrain, y_ktest = y_bin.iloc[train_idx], y_bin.iloc[test_idx]

            modelo_k = LogisticRegression(max_iter=1000)
            modelo_k.fit(X_ktrain, y_ktrain)
            y_kpred = modelo_k.predict(X_ktest)

            acc = accuracy_score(y_ktest, y_kpred)
            acuracias_kfold.append(acc)

        print(f" Acurácias por fold: {[f'{a*100:.2f}%' for a in acuracias_kfold]}")
        print(f" Acurácia média (KFold): {np.mean(acuracias_kfold) * 100:.2f}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# === RESUMO DO MODELO ========================================
print("\n📘 RESUMO DO MODELO")
print("""
Este modelo é um classificador binário que prevê, com base nas notas médias dos estudantes,
se as taxas educacionais de uma escola estarão altas ou baixas (acima da mediana nacional).
As taxas previstas são:
  - Abandono Escolar
  - Permanência Escolar
  - Aprovação
  - Reprovação

As escolas são classificadas como: Federal, Estadual, Privada ou Rural.
As notas utilizadas para previsão incluem: Matemática, Redação, Linguagens, etc.

Modelo usado: MLPClassifier (rede neural simples)
""")
# ============================================================

# Carregar os dados
df = pd.read_csv("MICRODADOS_ENEM_ESCOLA.csv", sep=';', encoding='latin-1', low_memory=False)

# Colunas
medias = ['NU_MEDIA_CN', 'NU_MEDIA_CH', 'NU_MEDIA_LP', 'NU_MEDIA_MT', 'NU_MEDIA_RED']
taxas = ['NU_TAXA_ABANDONO', 'NU_TAXA_PERMANENCIA', 'NU_TAXA_APROVACAO', 'NU_TAXA_REPROVACAO']

# Função para classificar tipo de escola
def classificar_escola(row):
    if row['TP_LOCALIZACAO_ESCOLA'] == 2:
        return 'Rural'
    elif row['TP_LOCALIZACAO_ESCOLA'] == 1:
        if row['TP_DEPENDENCIA_ADM_ESCOLA'] == 1:
            return 'Federal'
        elif row['TP_DEPENDENCIA_ADM_ESCOLA'] == 2:
            return 'Estadual'
        elif row['TP_DEPENDENCIA_ADM_ESCOLA'] == 4:
            return 'Privada'
    return 'Outros'

# Aplicar classificação
df['TIPO_ESCOLA'] = df.apply(classificar_escola, axis=1)
tipos_escola = ['Federal', 'Estadual', 'Privada', 'Rural']

# Lista para salvar resultados
resultados_grafico = []

# Avaliação por tipo de escola
for tipo in tipos_escola:
    df_tipo = df[df['TIPO_ESCOLA'] == tipo]
    df_tipo = df_tipo[medias + taxas].dropna()

    if df_tipo.empty:
        print(f"\n❌ Sem dados suficientes para escola do tipo: {tipo}")
        continue

    print(f"\n=====  MLPClassifier - {tipo} =====")
    X = df_tipo[medias]

    for taxa in taxas:
        y = df_tipo[taxa]
        y_bin = (y > y.median()).astype(int)

        print(f"\n Prevendo taxa: {taxa} (alta ou baixa)")

        # Separar treino e teste
        X_train, X_test, y_train, y_test = train_test_split(
            X, y_bin, test_size=0.2, random_state=42, stratify=y_bin
        )

        modelo = MLPClassifier(max_iter=1000, random_state=42)
        modelo.fit(X_train, y_train)
        y_pred = modelo.predict(X_test)
        acc_split = accuracy_score(y_test, y_pred)
        print(f" Acurácia (train/test split): {acc_split * 100:.2f}%")

        resultados_grafico.append({
            'Tipo': tipo,
            'Taxa': taxa.replace("NU_TAXA_", "").capitalize(),
            'Acuracia (%)': round(acc_split * 100, 2)
        })

# Plotar resultados
df_resultados = pd.DataFrame(resultados_grafico)
plt.figure(figsize=(12, 6))
for taxa in df_resultados['Taxa'].unique():
    subset = df_resultados[df_resultados['Taxa'] == taxa]
    plt.plot(subset['Tipo'], subset['Acuracia (%)'], marker='o', label=taxa)

plt.title(" Acurácia do MLPClassifier por Tipo de Escola e Taxa")
plt.xlabel("Tipo de Escola")
plt.ylabel("Acurácia (%)")
plt.legend(title="Taxa Avaliada")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
